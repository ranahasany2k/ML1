```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
%matplotlib inline
```

```python
df=pd.read_csv('Salary Data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
df.columns
```

```python
plt.xlabel("year_of_experience")
plt.ylabel("Salary(Rs)")
plt.scatter(df.YearsExperience,df.Salary,color="black",marker="*")
```

```python
x=df.drop("Salary",axis=1)
y=df.drop("YearsExperience",axis=1)
print(x.shape)
print(y.shape)
x1=df['YearsExperience']
y1=df['Salary']
print(x1.shape)
print(y1.shape)
x2=df.YearsExperience
y2=df.Salary
print(x2.shape)
print(y2.shape)
```

```python
reg = linear_model.LinearRegression()
reg.fit(x,y)
val=4.1
reg.predict([[val]])
```

```python
m=reg.coef_
print(m)
```

```python
c=reg.intercept_
print(c)
```

```python
y=m*val+c
print(y)
```

```python
plt.scatter(x, y, label='Data points')

# Plot the regression line
plt.plot(x, y, color='red', label='Regression line')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.show()
```

```python
plt.scatter(x, y, label='Data points')

# Plot the regression line
plt.plot(x, m * x + c, color='red', label='Regression line')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.show()
```

```python
sns.regplot(x=x, y=y,scatter_kws={'color': 'blue'}, line_kws={'color': 'red'}, label='Data points and Regression line')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
reg1 = linear_model.LinearRegression()
reg1.fit(X_train,y_train)
val=4.1
reg1.predict([[val]])
```

```python
y_prid=reg1.predict(X_test)
print(y_prid)
```

```python
sns.residplot(x=y_prid, y=y_test - y_prid, color="g")
plt.xlabel("Predicted values")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

```python
score=r2_score(y_test,y_prid)
print(score)
```

```python
mse = mean_squared_error(y_test, y_prid)
mae = mean_absolute_error(y_test, y_prid)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
```

```python
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.scatter(X_test, y_prid, color='red', label='Predict')
# plt.plot(X_test, y_prid, color='red', linewidth=2, label='Predicted')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Actual vs. Predicted Output')
plt.legend()
plt.show()
```
