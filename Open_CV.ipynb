```python
import cv2 as cv
import numpy as np
path='img.png'
image=cv.imread(path)

def playwithimage(image):
    rows,cols,channels= image.shape
    for i in range(360):
        M = cv.getRotationMatrix2D((cols/2,rows/2),i,1) 
        rotate = cv.warpAffine(image,M,(cols,rows)) 
    # cv.imshow('hello',rotate)
    # cv.waitKey(500)
    # cv.destroyAllWindows()
        alpha = 1
        beta = 80
        adjusted = cv.convertScaleAbs(rotate, alpha=alpha, beta=beta)
        noisy_image = cv.GaussianBlur(rotate, (7, 7), 0) 
        output_a=  f'rotate_brighted_noised{i}.png'
        output=  f'rotate_brighted{i}.png'
        output_filename = f'rotate_{i}.png'
        cv.imwrite(output,adjusted)
        cv.imwrite(output_filename,rotate)
        cv.imwrite(output_a,noisy_image)
playwithimage(image)
```
