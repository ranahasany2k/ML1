```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as file:
        # Create a PDF reader object
        pdf_reader = PyPDF2.PdfReader(file)

        # Iterate through all pages and extract text
        for page_number in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_number]
            text += page.extract_text()

    return text

# Example usage
pdf_path = "meta.pdf"
extracted_text = extract_text_from_pdf(pdf_path)

# Display or store the extracted text as needed
print(extracted_text)
```

```python
import re

def find_pattern_indexes(pattern, text):
    # Update the pattern with re.IGNORECASE and re.DOTALL flags
    compiled_pattern = re.compile(pattern, re.IGNORECASE | re.DOTALL)
    indexes = [match.start() for match in re.finditer(compiled_pattern, text)]
    return indexes

# Example usage
pattern_to_search = r'Cash and Cash Equivalents'
indexes_found = find_pattern_indexes(pattern_to_search, extracted_text)

# Display the indexes where the pattern is present
print("Indexes where the pattern is present:", indexes_found)
```

```python
for i in indexes_found:
    print(extracted_text[i:i+100])
```
