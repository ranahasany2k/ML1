```python
df["created_at"] = pd.to_datetime(df["created_at"], errors='coerce')
```

```python
plt.scatter(df.created_at, df.last_price, label='Scatter Plot', color='blue', marker='o')
```

```python
df.isnull().sum()
df.columns[df.isnull().sum() > 0].tolist()
df.fillna(df.mean(), inplace=True)
```
