```python
import cv2
from pathlib import Path

def read_coordinates(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        listt=[list(map(float, line.strip().split()[1:])) for line in lines]
        print(listt)
        return [list(map(float, line.strip().split()[1:])) for line in lines]

folder_path = Path('path/to/your/images/')

for i in range(1, 8):
    image_path = f'C:\\Users\\My computer & laptop\\Desktop\\Data Science\\ML1_tasks\\cat_dog\\{i}.jpeg'
    image = cv2.imread(str(image_path))

    coordinates_path = f'{i}.txt'
    coordinates = read_coordinates(str(coordinates_path))

    for coord in coordinates:
            x, y, w, h = map(float, coord)
            print(f"Drawing rectangle at ({x}, {y}), ({x + w}, {y + h}) for image {i}")
            cv2.rectangle(image, (int(x * image.shape[1]*0.8), int(y * image.shape[0]*0.65)),
                    (int((x + w) * image.shape[1]*0.8), int((y + h) * image.shape[0]*0.65)), (0, 255, 0), 2)

    cv2.imshow(f'Image {i} with Rectangles', image)
    cv2.waitKey(0)

cv2.destroyAllWindows()

```
